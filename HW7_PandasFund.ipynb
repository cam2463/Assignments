# 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "http://www.ldeo.columbia.edu/~danielmw/usgs_earthquakes_2014.csv"

df = pd.read_csv(url)

print(df.head())
print(df.info())
```

# 3

```python
stats = df.describe()
print(stats)

print(f"Max Magnitude: {df['mag'].max()}")
print(f"Min Magnitude: {df['mag'].min()}")
```

# 4

```python
top_20 = df.nlargest(20, 'mag')
print(top_20[['place', 'mag']])
```

# 5

```python
df['country'] = df['place'].str.split(',').str[-1].str.strip()
```

# 6

```python
print(df['country'].unique())
```

# 7

```python
filtered_df = df[df['mag'] >= 4]
```

# 8

```python
top_locations = filtered_df['country'].value_counts().head(5)

top_locations.plot(kind='bar', color='skyblue')
plt.title('Top 5 Locations for Earthquakes (Mag >= 4)')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```

# 9

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Unfiltered 
df['mag'].hist(ax=axes[0], bins=30, log=True, color='gray')
axes[0].set_title('All Earthquakes (Log Scale)')

# Filtered 
filtered_df['mag'].hist(ax=axes[1], bins=30, log=True, color='orange')
axes[1].set_title('Earthquakes Mag >= 4 (Log Scale)')

plt.show()
```

# 10 

```python
# Plot 1: Unfiltered Data
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], c=df['mag'], cmap='viridis', s=1, alpha=0.5)
plt.colorbar(label='Magnitude')
plt.title('Global Earthquake Distribution (Unfiltered)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Plot 2: Filtered Data (Mag >= 4)
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['longitude'], filtered_df['latitude'], 
            c=filtered_df['mag'], cmap='hot', s=10, alpha=0.7)
plt.colorbar(label='Magnitude')
plt.title('Significant Earthquakes (Magnitude >= 4)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Both maps show a global distribution of earhquakes but the filtered map shows that the earthquakes with the highest magnitude occur along tectonic plates
