#### 1)

```python
def convert_k_c(T, c = False):
    #unit = 0
    if c: #celcius to kelvin
        return T, "Celcius"  #+ 273.15 #+ unit
    else: #kelvin to celcius
        #unit = "K"
        return T - 273.15, "K" #+ unit
```

```python
convert_k_c(280) #kelvin to celcius
```

```python
convert_k_c(23, c = True) # celcius (c=True) to kelvin
```

#### 2)

```python
def convertToF(T, k = False):
    if k:
        T = T-273.15
        T = (T*(9/5)) + 32
        return T
    else:
        T = (T*(9/5)) + 32
        return T
    
```

#### 3)

```python
convertToF(100) #100C to F
```

```python
convertToF(274, k = True) #274K (k=true) to F
```

```python
def convertFromF(T,k = False):
    if k:
        T = (T-32)*5/9
        T = T + 273.15
        return T
    else:
        T = (T-32)*5/9
        return T
```

#### 5)

```python
def imp_to_meters(feet, inches):
  ft_to_m = feet * 0.3048
  in_to_m = inches * 0.0254
  meters = ft_to_m + in_to_m
  return meters
```

```python
imp_to_meters(6,1)
```

#### 6)

```python
def meters_to_imp(meters):
    total_ft = meters * 3.2808
    feet = int(total_ft)
    inches = (total_ft - feet) * 12
    return feet, inches
```
