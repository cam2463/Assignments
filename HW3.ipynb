```python
sample_json = '''
{
  "station": "USC00305800",
  "name": "New York Central Park",
  "location": {
    "latitude": 40.7789,
    "longitude": -73.9692
  },
  "observations": [
    {"date": "2023-01-01", "temperature": 32, "precipitation": 0.0},
    {"date": "2023-01-02", "temperature": 28, "precipitation": 0.5},
    {"date": "2023-01-03", "temperature": 35, "precipitation": 0.0},
    {"date": "2023-01-04", "temperature": 38, "precipitation": 0.2},
    {"date": "2023-01-05", "temperature": 41, "precipitation": 0.0}
  ]
}
'''
```

```python
data = json.loads(sample_json)
```

```python
print("Station:", data['station'])
```

```python
print("Location:", data['location'])
```

```python
print("First observation:", data['observations'][0])
```

## Task 1

```python
print("Date, Temperature")
for obs in data["observations"]: print(obs['date'], obs["temperature"])
```

## Task 2

```python
total_temp = 0 
count = 0
```

```python
for obs in data[ 'observations']:
    total_temp += obs ['temperature']
    count+=1
```

```python
average_temp = total_temp / count
print(f"The average temperature is: {average_temp}")
```

```python
print("Getting real weather data")
url = "https://api.weather.gov/points/40.7789,-73.9692"
response = requests.get(url)
if response.status_code == 200:
    weather_data = response.json()
    city = weather_data['properties'] ['relativeLocation'][ 'properties'][' city']
    state = weather_data['properties'][' relativeLocation' ]['properties' ]['state' ]
    print (f"Location: {city}, {state}")
    
    forecast_url = weather_data['properties']['forecast']
    forecast_response = requests.get(forecast_url)
    forecast = forecast_response.json()
   
    print("\nForecast:")
    print("-" * 50)
    for period in forecast["properties"]["periods"][:3]:
        print(f"{period['name']}:")
        print(f"Temperature: {period['temperature']}Â°{period["temperatureUnit"]}")
        print(f"Forecast: {period['shortForecast']}")
else:
    print(f"Error: {response.status_code}")
```
